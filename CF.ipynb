```python
cf = pd.DataFrame()
# df_mac["Subject"] = df.keys()
cf = pd.DataFrame(columns = ["Subject",
                             "Age",
                             "Gender",
                             "Genotype",
                             "Modulator",
                            "Repeated trials",
                            "Macroduct Concentration 1 (mM)","Macroduct Concentration 2 (mM)",
                            "Macroduct Concentration 3 (mM)","Macroduct Concentration 4 (mM)",
                            "Mac 1 - Mac 2",
                            "Mac 3 - Mac 4"],
                     index = df.keys())
```

```python
for key in df:
    cf.loc[key] = [key[0:4],
                   df[key].iat[0,2],
                   df[key].iat[0,3],
                   df[key].iat[0,5],
                   df[key].iat[0,4],
                    key[-2:],
                    df[key].iat[0,18],
                    df[key].iat[0,19],
                    df[key].iat[0,32],
                    df[key].iat[0,33],
                    df[key].iat[0,18] - df[key].iat[0,19],
                    df[key].iat[0,32] - df[key].iat[0,33]]
cf = cf.drop("template")
```

```python
# get a set of genotype using for sorting
set(cf["Genotype"])
```

```python
cf["Genotype"] = pd.Categorical(cf["Genotype"], ordered = True, 
                           categories = ["F508del/F508del", 
                                         "F508del/G551D", 
 'F508del / 3849+10kbC>T',
 'F508del/2789+5GA',
 'F508del/3905insT',
 'F508del/G542X',
 'F508del/R1066C',
 'F508del/W1282X',
 'G551D/2622+1G>A',
 'W1282X/2789+5G -> A'])
```

```python
cf_sorted = cf.sort_values(by = ["Genotype", "Subject"], ascending = True)
```

```python
mapping = {item:i for i, item in enumerate(cf_sorted["Subject"].unique(),1)}
# insert column at a specific location
cf_sorted.insert(loc = 1, column = 'Subject Number', value = cf_sorted["Subject"].apply(lambda x: mapping[x]))
# cf_sorted["Subject Number"] = cf_sorted["Subject"].apply(lambda x: mapping[x])
```
