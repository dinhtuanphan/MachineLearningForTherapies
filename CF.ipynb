```python
import pandas as pd
df = pd.read_excel('20210518_updated_CF_subjects_iontophoresis.xlsx', sheet_name = None)
```

```python
subjects = set()
for key in df.keys():
    subjects.add(key[0:4])    
print(subjects)
```

```python
subjects.remove("temp")
len(subjects)
```

```python
cf = pd.DataFrame()
# df_mac["Subject"] = df.keys()
cf = pd.DataFrame(columns = ["Subject",
                             "Age",
                             "Gender",
                             "Genotype",
                             "Modulator",
                            "Repeated trials",
                            "Macroduct Concentration 1 (mM)","Macroduct Concentration 2 (mM)",
                            "Macroduct Concentration 3 (mM)","Macroduct Concentration 4 (mM)",
                            "Mac 1 - Mac 2",
                            "Mac 3 - Mac 4"],
                     index = df.keys())
```

```python
for key in df:
    cf.loc[key] = [key[0:4],
                   df[key].iat[0,2],
                   df[key].iat[0,3],
                   df[key].iat[0,5],
                   df[key].iat[0,4],
                    key[-2:],
                    df[key].iat[0,18],
                    df[key].iat[0,19],
                    df[key].iat[0,32],
                    df[key].iat[0,33],
                    df[key].iat[0,18] - df[key].iat[0,19],
                    df[key].iat[0,32] - df[key].iat[0,33]]
cf = cf.drop("template")
```

```python
# get a set of genotype using for sorting
set(cf["Genotype"])
```

```python
replace_values = {'F508del / 3849+10kbC>T':'F508del/3849+10kbC>T', 
                 'F508del/2789+5GàA':'F508del/2789+5G>A',
                 'W1282X/2789+5G -> A':'W1282X/2789+5G>A'}
cf.replace({'Genotype':replace_values}, inplace=True)
```

```python
cf_grp = cf.groupby('Genotype', as_index = False, sort = True)['Subject'].count().sort_values(by='Subject', ascending = False)
cf_grp
```

```python
cf["Genotype"] = pd.Categorical(cf["Genotype"], ordered = True, 
                           categories = cf_grp['Genotype'])
```

```python
cf_sorted = cf.sort_values(by = ["Genotype", "Subject"], ascending = True)
```

```python
mapping = {item:i for i, item in enumerate(cf_sorted["Subject"].unique(),1)}
# insert column at a specific location
cf_sorted.insert(loc = 1, column = 'Subject Number', value = cf_sorted["Subject"].apply(lambda x: mapping[x]))
# cf_sorted["Subject Number"] = cf_sorted["Subject"].apply(lambda x: mapping[x])
```
